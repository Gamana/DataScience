# Sales Data Visualization — Questions & Solutions

Use the **sales dataset** (`sales_data.csv`) — **20 records**, **6 columns** — to create and interpret plots with **Matplotlib** and **Seaborn**.

| Column | Description |
|--------|-------------|
| `Order_ID` | Unique order identifier |
| `Date` | Date of the sale |
| `Product` | Laptop, Phone, Tablet, Monitor |
| `Region` | North, South, East, West |
| `Units_Sold` | Number of units sold |
| `Revenue` | Total revenue in dollars |

```python
%pip install matplotlib seaborn pandas
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sales_df = pd.read_csv('sales_data.csv')
sales_df['Date'] = pd.to_datetime(sales_df['Date'])
sales_df.head()
```

---
## Part 1 — Matplotlib Plots

---
### **Before the Plot**
### **Graph 1:** Line Plot — Total Revenue Over Time

**Questions to create this graph:**
- How does total revenue change from one sale date to another?
- Which month shows the highest single-day revenue?
- Is there an upward or downward trend in sales over the first half of 2024?

**Why is this graph used?**
A **line plot** is ideal for showing **trends over time**. It helps visualize how revenue fluctuates across different order dates.

```python
# Solution — Graph 1: Line Plot
plt.figure(figsize=(10, 5))
plt.plot(sales_df['Date'], sales_df['Revenue'], marker='o', linestyle='-', color='steelblue')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.title('Line Plot — Revenue by Order Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 1:** Line Plot — Total Revenue Over Time

**Questions to answer from this graph:**
- On which dates did revenue peak above $7,000?
- Are there noticeable dips between consecutive orders?
- Does revenue generally increase toward June 2024?

**Answers:**
- Revenue peaked above $7,000 on **Feb 15, 2024** ($8,000) and **May 2, 2024** ($7,200).
- Yes — revenue drops noticeably after peaks (e.g., $8,000 → $2,000 in March, $7,200 → $2,500 in May).
- No clear steady increase; revenue **fluctuates** month to month rather than trending upward.

---
### **Before the Plot**
### **Graph 2:** Bar Chart — Total Revenue by Region

**Questions to create this graph:**
- Which region generates the most total revenue?
- How does revenue compare across North, South, East, and West?
- Are there regions that underperform compared to others?

**Why is this graph used?**
A **bar chart** compares **categorical groups**. It makes it easy to rank regions by their total sales performance.

```python
# Solution — Graph 2: Bar Chart
revenue_by_region = sales_df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(revenue_by_region.index, revenue_by_region.values, color=['#4C72B0', '#55A868', '#C44E52', '#8172B2'])
plt.xlabel('Region')
plt.ylabel('Total Revenue ($)')
plt.title('Bar Chart — Total Revenue by Region')
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 2:** Bar Chart — Total Revenue by Region

**Questions to answer from this graph:**
- Which region has the highest and lowest total revenue?
- Is the revenue gap between the top two regions significant?
- Which region should the sales team focus on for growth?

**Answers:**
- **Highest:** East ($23,700) | **Lowest:** West ($15,200).
- The gap between East and North is only **$1,800** — not very large; top three regions are close.
- Focus on **West** for growth — it has the lowest total revenue and missed Phone sales entirely.

---
### **Before the Plot**
### **Graph 3:** Pie Chart — Revenue Share by Product

**Questions to create this graph:**
- What percentage of total revenue comes from each product?
- Which product contributes the largest share of sales?
- Are any products barely contributing to overall revenue?

**Why is this graph used?**
A **pie chart** shows **proportions of a whole**. It is useful when you want to see how revenue is divided among product categories.

```python
# Solution — Graph 3: Pie Chart
revenue_by_product = sales_df.groupby('Product')['Revenue'].sum()

plt.figure(figsize=(7, 7))
plt.pie(revenue_by_product.values, labels=revenue_by_product.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart — Revenue Share by Product')
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 3:** Pie Chart — Revenue Share by Product

**Questions to answer from this graph:**
- Which product holds the largest revenue share?
- Do Laptop and Phone together account for more than half of total revenue?
- Should the company invest more in the lowest-share product?

**Answers:**
- **Phone** holds the largest share at ~39% ($32,000 of $82,100).
- Yes — Laptop + Phone together account for **~64.6%** of total revenue.
- **Tablet** has the smallest share (~16.4%); investment depends on margin and growth potential, not share alone.

### **Before the Plot**
### **Graph 4:** Scatter Plot — Units Sold vs Revenue

**Questions to create this graph:**
- Is there a relationship between units sold and revenue per order?
- Do orders with more units always generate higher revenue?
- Are there outliers where few units still produced high revenue?

**Why is this graph used?**
A **scatter plot** reveals the **relationship between two numeric variables**. It helps identify patterns, clusters, and outliers in sales data.

```python
# Solution — Graph 4: Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(sales_df['Units_Sold'], sales_df['Revenue'], color='coral', s=80, alpha=0.8)
plt.xlabel('Units Sold')
plt.ylabel('Revenue ($)')
plt.title('Scatter Plot — Units Sold vs Revenue')
plt.grid(True, alpha=0.3)
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 4:** Scatter Plot — Units Sold vs Revenue

**Questions to answer from this graph:**
- Does revenue generally increase as units sold increase?
- Which orders are outliers (high revenue with relatively few units)?
- What might explain orders that deviate from the overall pattern?

**Answers:**
- Within each product type, more units tend to mean higher revenue, but points form **clusters** by product (different price per unit).
- **Laptop orders** (e.g., Order 1009: 4 units → $6,000) earn high revenue with fewer units than Phone orders.
- Different **products have different unit prices** — Laptops ~$1,500/unit, Phones ~$800/unit, Tablets/Monitors ~$500–600/unit.

### **Before the Plot**
### **Graph 5:** Histogram — Distribution of Revenue

**Questions to create this graph:**
- How is revenue spread across all 20 orders?
- Are most orders clustered around a specific revenue range?
- Are there unusually high or low revenue orders?

**Why is this graph used?**
A **histogram** shows the **distribution** of a single numeric variable. It helps understand how frequently different revenue levels occur.

```python
# Solution — Graph 5: Histogram
plt.figure(figsize=(8, 5))
plt.hist(sales_df['Revenue'], bins=6, color='mediumseagreen', edgecolor='black')
plt.xlabel('Revenue ($)')
plt.ylabel('Number of Orders')
plt.title('Histogram — Distribution of Order Revenue')
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 5:** Histogram — Distribution of Revenue

**Questions to answer from this graph:**
- What is the most common revenue range for orders?
- Is the distribution skewed toward lower or higher values?
- How many orders fall above $6,000 in revenue?

**Answers:**
- Most orders cluster in the **$2,500–$4,500** range.
- The distribution is slightly **right-skewed** — a few high-revenue Phone orders pull the tail upward.
- **3 orders** exceed $6,000 (Orders 1002, 1006, and 1014).

## Part 2 — Seaborn Plots

---
### **Before the Plot**
### **Graph 6:** Seaborn Bar Plot — Average Revenue by Product

**Questions to create this graph:**
- Which product has the highest average revenue per order?
- How does average order value differ between Laptop, Phone, Tablet, and Monitor?
- Which product generates the most revenue per transaction on average?

**Why is this graph used?**
Seaborn's `barplot()` computes and displays **group averages** with clean styling, making product comparisons easy to read.

```python
# Solution — Graph 6: Seaborn Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_df, x='Product', y='Revenue', estimator='mean', palette='viridis')
plt.xlabel('Product')
plt.ylabel('Average Revenue ($)')
plt.title('Seaborn Bar Plot — Average Revenue by Product')
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 6:** Seaborn Bar Plot — Average Revenue by Product

**Questions to answer from this graph:**
- Which product has the highest and lowest average revenue per order?
- Is Phone's average revenue noticeably higher than Tablet's?
- Based on this chart, which product is the most profitable per sale?

**Answers:**
- **Highest:** Phone ($6,400 avg) | **Lowest:** Tablet ($2,700 avg).
- Yes — Phone averages **$3,700 more** per order than Tablet.
- **Phone** is the most profitable per transaction on average.

**Questions to create this graph:**

### **Before the Plot**
### **Graph 8:** Seaborn Box Plot — Revenue Distribution by Product

**Questions to create this graph:**
- How does revenue vary within each product category?
- Which product shows the widest spread in revenue values?
- Are there any unusual (outlier) revenue values for a given product?

**Why is this graph used?**
A **box plot** shows the **median, spread, and outliers** of a numeric variable grouped by category — useful for comparing revenue consistency across products.

```python
# Solution — Graph 8: Seaborn Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df, x='Product', y='Revenue', palette='pastel')
plt.xlabel('Product')
plt.ylabel('Revenue ($)')
plt.title('Seaborn Box Plot — Revenue by Product')
plt.show()
```

---

### **Graph 8:** Seaborn Box Plot — Revenue Distribution by Product

**Questions to answer from this graph:**
- Which product has the highest median revenue?
- Which product shows the most variation in order revenue?
- Do any products have outlier orders far from the typical range?

**Answers:**
- **Phone** has the highest median revenue ($6,400).
- **Phone** shows the widest spread ($4,800–$8,000), while **Tablet** is the most consistent ($2,000–$3,500).
- No extreme outliers — all products have relatively tight ranges within their price tiers.

### **Before the Plot**
### **Graph 9:** Seaborn Heatmap — Total Revenue by Region and Product

**Questions to create this graph:**
- Which region–product combination generates the most revenue?
- Are there region–product pairs with very low sales?
- Can we spot patterns such as Laptops selling best in the North?

**Why is this graph used?**
A **heatmap** displays values in a table using color intensity. It is excellent for comparing two categorical dimensions (Region × Product) at a glance.

```python
# Solution — Graph 9: Seaborn Heatmap
pivot_table = sales_df.pivot_table(index='Region', columns='Product', values='Revenue', aggfunc='sum', fill_value=0)

plt.figure(figsize=(8, 5))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlOrRd', linewidths=0.5)
plt.title('Seaborn Heatmap — Revenue by Region and Product')
plt.ylabel('Region')
plt.xlabel('Product')
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 9:** Seaborn Heatmap — Total Revenue by Region and Product

**Questions to answer from this graph:**
- Which single region–product cell has the highest revenue?
- Which combinations have zero or very low revenue?
- Should marketing focus on weak region–product pairs?

**Answers:**
- **East + Phone** generates the most revenue ($15,200).
- Several cells are **$0**: North/Tablet, South/Laptop, West/Phone — no sales for those pairs.
- Yes — target **West/Phone** and **North/Tablet** where no sales occurred despite demand in other regions.

### **Before the Plot**
### **Graph 10:** Seaborn Line Plot — Monthly Total Revenue Trend

**Questions to create this graph:**
- How does total monthly revenue change from January to June 2024?
- Which month had the highest combined sales?
- Is there a seasonal pattern in monthly revenue?

**Why is this graph used?**
Seaborn's `lineplot()` is well-suited for showing **aggregated trends over time**, with smooth styling and optional confidence intervals.

```python
# Solution — Graph 10: Seaborn Line Plot
sales_df['Month'] = sales_df['Date'].dt.to_period('M').astype(str)
monthly_revenue = sales_df.groupby('Month', as_index=False)['Revenue'].sum()

plt.figure(figsize=(9, 5))
sns.lineplot(data=monthly_revenue, x='Month', y='Revenue', marker='o', linewidth=2.5, color='darkorange')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.title('Seaborn Line Plot — Monthly Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

---
### **After the Plot — Review & Solutions**
### **Graph 10:** Seaborn Line Plot — Monthly Total Revenue Trend

**Questions to answer from this graph:**
- Which month recorded the highest total revenue?
- Did revenue grow steadily or fluctuate month to month?
- What business decisions could be made based on this monthly trend?

**Answers:**
- **March 2024** had the highest monthly revenue ($15,400).
- Revenue **fluctuated** — it dipped in April ($10,500) before recovering in May.
- Investigate the **April dip**, replicate March's success, and plan inventory ahead of peak months.
