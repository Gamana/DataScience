#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

student_df = pd.read_csv('../pandas/studentsample.csv')
```

```python
### Basic Plotting With Seaborn
student_df
```

---
### **Before the Plot**
### **Graph Name:** Scatter Plot — Age vs GPA

**Questions to create this graph:**
- Is there a relationship between student age and GPA?
- Do older students tend to have higher or lower GPAs?
- Are there outliers where GPA is unusually high or low for a given age?

**Why is this graph used?**
A **scatter plot** shows the relationship between two numeric variables. Seaborn's `scatterplot` makes it easy to visualize whether **age and academic performance** are related across students.

```python
##create a scatter plot
scatter_df = student_df.dropna(subset=['age', 'gpa'])

sns.scatterplot(x='age', y='gpa', data=scatter_df)
plt.title("Scatter Plot of Age vs GPA")
plt.xlabel('Age')
plt.ylabel('GPA')
plt.show()
```

### **Graph Name:** Scatter Plot — Age vs GPA

**Why this graph was used:**
A **scatter plot** shows the relationship between two numeric variables. Seaborn's `scatterplot` makes it easy to visualize whether **age and academic performance** are related across students.

### **Before the Plot**
### **Graph Name:** Line Plot — Average GPA by Enrollment Year

**Questions to create this graph:**
- How does average GPA change across enrollment years?
- Which year has the highest or lowest average GPA?
- Is academic performance improving or declining over time?

**Why is this graph used?**
A **line plot** is ideal for showing **trends over ordered categories or time**. Seaborn's `lineplot` connects summary values to highlight how average GPA changes across enrollment years.

```python
## Line Plot
gpa_by_year = student_df.groupby('enrollment_year', as_index=False)['gpa'].mean().dropna()

sns.lineplot(x='enrollment_year', y='gpa', data=gpa_by_year, marker='o')
plt.title("Line Plot of Average GPA by Enrollment Year")
plt.xlabel('Enrollment Year')
plt.ylabel('Average GPA')
plt.show()
```

---
### **After the Plot — Review**
### **Graph Name:** Line Plot — Average GPA by Enrollment Year

**Questions to answer from this graph:**
- Which enrollment year shows the highest average GPA?
- Is the GPA trend increasing, decreasing, or flat?
- What changes in performance can you infer year to year?

**Why this graph was used:**
A **line plot** is ideal for showing **trends over ordered categories or time**. Seaborn's `lineplot` connects summary values to highlight how average GPA changes across enrollment years.

### **Before the Plot**
### **Graph Name:** Bar Plot — Average GPA by Major

**Questions to create this graph:**
- Which major has the highest average GPA?
- Which major has the lowest average GPA?
- How do academic outcomes differ across majors?

**Why is this graph used?**
A **bar plot** compares a numeric summary across categories. Seaborn's `barplot` is useful for comparing **average GPA across majors** in a clear, ranked format.

```python
## Categorical Plots
## Bar Plot
bar_df = student_df.dropna(subset=['major', 'gpa'])

sns.barplot(x='major', y='gpa', data=bar_df, estimator='mean')
plt.title('Bar Plot of Average GPA by Major')
plt.xlabel('Major')
plt.ylabel('Average GPA')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### **Before the Plot**
### **Graph Name:** Box Plot — GPA by Gender

**Questions to create this graph:**
- How is GPA distributed for male vs female students?
- Which gender group has a higher median GPA?
- Are there more outliers in one gender group than the other?

**Why is this graph used?**
A **box plot** summarizes distribution with median, quartiles, and outliers. Seaborn's `boxplot` helps compare **GPA spread and central tendency** between gender groups.

```python
## Box Plot
box_df = student_df.dropna(subset=['gender', 'gpa'])

sns.boxplot(x='gender', y='gpa', data=box_df)
plt.title('Box Plot of GPA by Gender')
plt.xlabel('Gender')
plt.ylabel('GPA')
plt.show()
```

### **Before the Plot**
### **Graph Name:** Violin Plot — GPA by Major

**Questions to create this graph:**
- How is GPA distributed within each major?
- Which majors show the widest GPA spread?
- Do some majors have multiple concentration points in GPA?

**Why is this graph used?**
A **violin plot** combines a box plot with a **density shape**, showing both summary stats and distribution form. It is useful for comparing **GPA distributions across majors** in more detail than a simple bar chart.

```python
## Violin Plot
violin_df = student_df.dropna(subset=['major', 'gpa'])

sns.violinplot(x='major', y='gpa', data=violin_df)
plt.title('Violin Plot of GPA by Major')
plt.xlabel('Major')
plt.ylabel('GPA')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### **Graph Name:** Histogram — GPA Distribution

**Questions to create this graph:**
- What is the most common GPA range among students?
- Is the GPA distribution skewed toward high or low values?
- How many students fall in each GPA interval?

**Why is this graph used?**
A **histogram** shows how numeric data is spread across bins. Seaborn's `histplot` with `kde=True` adds a smooth density curve, making it easier to understand the **overall GPA distribution**.

```python
### Histograms
sns.histplot(student_df['gpa'].dropna(), bins=5, kde=True)
plt.title('Histogram of GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Count')
plt.show()
```

### **Before the Plot**
### **Graph Name:** KDE Plot — Attendance Percentage

**Questions to create this graph:**
- What attendance percentage is most common among students?
- Is attendance concentrated in a narrow or wide range?
- Are there multiple peaks in attendance behavior?

**Why is this graph used?**
A **KDE (Kernel Density Estimate) plot** shows the **probability density** of a numeric variable as a smooth curve. It is useful for understanding the shape of **attendance percentage** without bin boundaries.

```python
## KDE Plot
sns.kdeplot(student_df['attendance_pct'].dropna(), fill=True)
plt.title('KDE Plot of Attendance Percentage')
plt.xlabel('Attendance (%)')
plt.ylabel('Density')
plt.show()
```

**Questions to answer from this graph:**
- Where does the density peak on the attendance scale?
- Is attendance mostly high, low, or mixed?
- Does the filled KDE suggest a single cluster or multiple clusters?

**Why this graph was used:**
A **KDE (Kernel Density Estimate) plot** shows the **probability density** of a numeric variable as a smooth curve. It is useful for understanding the shape of **attendance percentage** without bin boundaries.

**Questions to create this graph:**

**Why is this graph used?**
A **pairplot** creates a grid of scatter plots and distributions for multiple numeric variables. It is ideal for **exploratory analysis** when you want to inspect many student feature relationships together.
