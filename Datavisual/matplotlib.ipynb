#### Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
import pandas as pd

student_df = pd.read_csv('../pandas/studentsample.csv')
```

---
### **Before the Plot**
### **Graph Name:** Line Plot — Average GPA by Enrollment Year

**Questions to create this graph:**
- How does average GPA change across different enrollment years?
- Which enrollment year has the highest or lowest average GPA?
- Is there an upward or downward trend in academic performance over time?

**Why is this graph used?**
A **line plot** connects data points with lines, making it ideal for showing **trends over time** or across ordered categories. Here, it helps us visualize how student academic performance (GPA) varies from one enrollment year to another.

```python
gpa_by_year = student_df.groupby('enrollment_year')['gpa'].mean().dropna()

##create a line plot
plt.plot(gpa_by_year.index, gpa_by_year.values)
plt.xlabel('Enrollment Year')
plt.ylabel('Average GPA')
plt.title("Basic Line Plot - Average GPA by Enrollment Year")
plt.show()
```

### **Before the Plot**
### **Graph Name:** Multiple Subplots — GPA, Attendance & Student Count by Year

**Questions to create this graph:**
- How do GPA, attendance, and enrollment count compare across years?
- Do years with more students show different GPA or attendance patterns?
- Can we view multiple student metrics side by side in one figure?

**Why is this graph used?**
**Multiple subplots** display several related charts in a single figure. This is useful when comparing different aspects of the same dataset (GPA, attendance, student count) without switching between separate plots, helping spot relationships across metrics.

```python
## Multiple Plots
yearly = student_df.groupby('enrollment_year').agg(
    avg_gpa=('gpa', 'mean'),
    avg_attendance=('attendance_pct', 'mean'),
    student_count=('student_id', 'count')
).dropna()

years = yearly.index.astype(int)
y1 = yearly['avg_gpa']
y2 = yearly['avg_attendance']

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(years, y1, color='green')
plt.title("Average GPA by Year")

plt.subplot(2,2,2)
plt.plot(y1, years, color='red')
plt.title("GPA vs Enrollment Year")

plt.subplot(2,2,3)
plt.plot(years, y2, color='blue')
plt.title("Average Attendance by Year")

plt.subplot(2,2,4)
plt.plot(years, yearly['student_count'], color='green')
plt.title("Student Count by Year")
```

**Questions to create this graph:**
- Which majors have unusual GPA values that could be outliers?

**Why is this graph used?**
A **box plot** shows median, quartiles, and **outliers** in numeric data grouped by category. It is widely used in **data cleaning** to spot extreme GPA values that may need review, correction, or removal before analysis.

```python
## Box Plot — Outliers & Data Cleaning
box_df = student_df.dropna(subset=['major', 'gpa'])
majors = sorted(box_df['major'].unique())
box_data = [box_df.loc[box_df['major'] == major, 'gpa'].values for major in majors]

plt.figure(figsize=(10, 6))
plt.boxplot(box_data, labels=majors)
plt.xlabel('Major')
plt.ylabel('GPA')
plt.title('Box Plot — GPA Outliers by Major (Data Cleaning)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

### **Before the Plot**
### **Graph Name:** Stack Plot — Student Composition by Major

**Questions to create this graph:**
- How is the student population composed across majors each year?
- Which majors contribute most to total enrollment over time?
- How does the mix of majors change from year to year?

**Why is this graph used?**
A **stack plot** shows how parts combine to form a whole over an ordered axis. Like analyzing **revenue sources**, it helps study **composition** — here, how different majors contribute to total student enrollment across years.

```python
## Stack Plot — Composition Analysis
composition_df = student_df.dropna(subset=['enrollment_year', 'major'])
composition = composition_df.groupby(['enrollment_year', 'major']).size().unstack(fill_value=0)
years = composition.index.astype(int)

plt.stackplot(
    years,
    [composition[col] for col in composition.columns],
    labels=composition.columns,
    alpha=0.85
)
plt.xlabel('Enrollment Year')
plt.ylabel('Number of Students')
plt.title('Stack Plot — Student Composition by Major')
plt.legend(title='Major', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### **Before the Plot**
### **Graph Name:** Area Plot — Cumulative Enrollment Growth

**Questions to create this graph:**
- How does total student enrollment accumulate over time?
- Is enrollment growth steady or accelerating?
- What is the cumulative student count by the latest year?

**Why is this graph used?**
An **area plot** highlights a **cumulative trend** by filling the region under a line. It is useful for **growth analysis**, showing how total student enrollment builds up across enrollment years.

```python
## Area Plot — Cumulative Enrollment Growth
enrollment_counts = student_df.groupby('enrollment_year')['student_id'].count().sort_index()
cumulative_enrollment = enrollment_counts.cumsum()
years = cumulative_enrollment.index.astype(int)

plt.figure(figsize=(9, 5))
plt.fill_between(years, cumulative_enrollment.values, alpha=0.35, color='steelblue')
plt.plot(years, cumulative_enrollment.values, color='steelblue', linewidth=2, marker='o')
plt.xlabel('Enrollment Year')
plt.ylabel('Cumulative Number of Students')
plt.title('Area Plot — Cumulative Enrollment Growth')
plt.grid(True, alpha=0.3)
plt.show()
```

### **Before the Plot**
### **Graph Name:** Error Bar Plot — GPA Variability by Major

**Questions to create this graph:**
- What is the average GPA for each major?
- How much variability exists around the mean GPA in each major?
- Which majors show the most inconsistent academic performance?

**Why is this graph used?**
An **error bar plot** displays a central value plus **variability** (such as standard deviation). It is useful in **experiments** and comparisons to show not just the average GPA, but also how much results vary within each major.

```python
## Error Bar Plot — GPA Variability by Major
error_stats = student_df.dropna(subset=['major', 'gpa']).groupby('major')['gpa'].agg(['mean', 'std']).dropna()
x_pos = range(len(error_stats))

plt.figure(figsize=(10, 6))
plt.errorbar(
    x_pos,
    error_stats['mean'],
    yerr=error_stats['std'],
    fmt='o',
    capsize=6,
    color='darkorange',
    ecolor='gray',
    linewidth=2
)
plt.xticks(x_pos, error_stats.index, rotation=45, ha='right')
plt.xlabel('Major')
plt.ylabel('GPA')
plt.title('Error Bar Plot — GPA Variability by Major')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

### **Before the Plot**
### **Graph Name:** Subplots — Student Data Dashboard

**Questions to create this graph:**
- Can we view GPA, attendance, city, and scholarship insights in one dashboard?
- Which metrics should appear together for quick student analysis?
- How can multiple views help compare different aspects of the dataset?

**Why is this graph used?**
**Subplots** combine **multiple views** in one figure, making them ideal for **dashboards**. Here, four related student charts are shown together so you can monitor performance, attendance, location, and scholarship status at a glance.

```python
## Subplots — Student Data Dashboard
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Student Data Dashboard', fontsize=14, fontweight='bold')

# View 1: Average GPA by gender
gender_gpa = student_df.dropna(subset=['gender', 'gpa']).groupby('gender')['gpa'].mean()
axes[0, 0].bar(gender_gpa.index, gender_gpa.values, color=['salmon', 'skyblue'])
axes[0, 0].set_title('Avg GPA by Gender')
axes[0, 0].set_ylabel('GPA')

# View 2: Attendance distribution
axes[0, 1].hist(student_df['attendance_pct'].dropna(), bins=5, color='seagreen', edgecolor='black')
axes[0, 1].set_title('Attendance Distribution')
axes[0, 1].set_xlabel('Attendance (%)')

# View 3: Top cities by enrollment
city_counts = student_df['city'].value_counts().head(6)
axes[1, 0].barh(city_counts.index, city_counts.values, color='mediumpurple')
axes[1, 0].set_title('Top Cities by Enrollment')
axes[1, 0].set_xlabel('Students')

# View 4: Scholarship status
scholarship_counts = student_df['scholarship'].fillna('Unknown').value_counts()
axes[1, 1].pie(scholarship_counts.values, labels=scholarship_counts.index, autopct='%1.0f%%')
axes[1, 1].set_title('Scholarship Status')

plt.tight_layout()
plt.show()
```
