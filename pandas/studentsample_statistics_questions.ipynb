## Suggested Learning Path

1. **Section A** — Mean: central tendency with Pandas and NumPy.
2. **Section B** — Median: robust center for skewed data.
3. **Section C** — Mode: most frequent value (numeric and categorical).
4. **Section D** — Range: spread from min to max.
5. **Section E** — Variance: average squared deviation from the mean.
6. **Section F** — Standard deviation: square root of variance, same units as data.
7. **Section G** — Quartiles: divide sorted data into four equal parts.
8. **Section H** — IQR: middle 50% spread (Q3 − Q1).
9. **Section I** — Percentiles: position-based thresholds in a distribution.
