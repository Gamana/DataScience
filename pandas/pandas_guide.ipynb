## 1. Introduction and Object Creation

Pandas has two primary data structures:
*   **Series**: A 1D labeled array capable of holding any data type.
*   **DataFrame**: A 2D labeled data structure with columns of potentially different types (like a table).

### Key Methods/Functions covered:
*   `pd.Series(data, index)`: Creates a 1D Series.
*   `pd.DataFrame(data, index, columns)`: Creates a 2D DataFrame.
*   `pd.read_csv()` / `to_csv()`: Reading/Writing CSV files.
*   `pd.read_json()` / `to_json()`: Reading/Writing JSON files.


```python
# Creating a Series
series_data = [10, 20, 30, 40, 50]
labels = ['a', 'b', 'c', 'd', 'e']
s = pd.Series(data=series_data, index=labels)
print("Pandas Series:\n", s, "\n")

# Creating a DataFrame from a Dictionary
dict_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000],
    'City': ['New York', 'Paris', 'London', 'Tokyo']
}
df = pd.DataFrame(data=dict_data)
print("Pandas DataFrame:\n", df)

# Simulating Writing and Reading CSV files
df.to_csv('sample_data.csv', index=False)
loaded_df = pd.read_csv('sample_data.csv')
print("\nLoaded from CSV:\n", loaded_df)
```

## 2. Viewing and Inspecting Data

Before modeling or cleaning, you must inspect the data to understand its shape, columns, missingness, and types.

### Key Methods/Attributes covered:
*   `df.head(n)`: Returns the first `n` rows.
*   `df.tail(n)`: Returns the last `n` rows.
*   `df.sample(n)`: Returns a random sample of `n` rows.
*   `df.info()`: Prints a concise summary of the DataFrame (index, dtypes, non-null values, memory).
*   `df.describe()`: Generates descriptive statistics for numerical columns.
*   `df.shape`: Attribute returning `(rows, columns)`.
*   `df.dtypes`: Attribute returning the data type of each column.
*   `df.columns` / `df.index`: Access column headers and index labels.
*   `df.memory_usage()`: Returns memory usage of each column in bytes.


```python
# Head, Tail, and Sample
print("Head (First 2 rows):")
display(df.head(2))

print("\nTail (Last 2 rows):")
display(df.tail(2))

print("\nSample (1 random row):")
display(df.sample(1))

# Inspecting characteristics
print("\nDataFrame Shape:", df.shape)
print("\nDataFrame dtypes:\n", df.dtypes)
print("\nDataFrame Columns:", df.columns.tolist())
print("\nDataFrame Index:", df.index)

# Info summary
print("\n--- DataFrame Info ---")
df.info()

# Summary statistics
print("\nDescriptive Statistics:")
display(df.describe())

# Memory Usage
print("\nMemory Usage (bytes):\n", df.memory_usage())
```

## 3. Selection and Indexing

Selecting subsets of data is one of the most common tasks. Pandas provides label-based (`loc`) and integer-based (`iloc`) selection.

### Key Methods/Indexers covered:
*   `df['col_name']`: Column selection (returns a Series).
*   `df[['col1', 'col2']]`: Select multiple columns.
*   `df.loc[row_label, col_label]`: Access rows and columns by labels/boolean arrays.
*   `df.iloc[row_pos, col_pos]`: Access rows and columns by integer positions.
*   `df[df['col'] > val]`: Boolean indexing / filtering.
*   `df.query('expression')`: Filter rows using a query string.
*   `df.set_index('col')` / `df.reset_index()`: Set a column as index, or reset it.


```python
# Selection by Column name
print("Select single column (Name):")
display(df['Name'])

# Selection using loc (label-based)
print("\nSelecting Alice and Bob (rows 0 and 1) with 'Name' and 'Salary' using .loc:")
display(df.loc[0:1, ['Name', 'Salary']])

# Selection using iloc (position-based)
print("\nSelecting first 2 rows and first 2 columns using .iloc:")
display(df.iloc[0:2, 0:2])

# Boolean indexing
print("\nFilter: Age > 30:")
display(df[df['Age'] > 30])

# Using .query()
print("\nFilter using query: City == 'New York' or Salary > 65000:")
display(df.query("City == 'New York' or Salary > 65000"))

# Setting and resetting index
print("\nSetting 'Name' as index:")
df_indexed = df.set_index('Name')
display(df_indexed)

print("\nResetting index back to default integers:")
display(df_indexed.reset_index())
```

## 4. Data Cleaning and Handling Missing Values

Real-world datasets contain nulls, duplicates, and poorly formatted column names.

### Key Methods covered:
*   `df.isnull()` / `df.isna()`: Check for missing values (returns boolean DataFrame).
*   `df.notnull()`: Opposite of `isnull()`.
*   `df.dropna()`: Drops rows or columns containing missing values.
*   `df.fillna(value)`: Replaces missing values with a specified value.
*   `df.replace(old, new)`: Replaces specific values with other values.
*   `df.duplicated()`: Identifies duplicate rows.
*   `df.drop_duplicates()`: Removes duplicate rows.
*   `df.rename(columns={'old': 'new'})`: Renames columns or indexes.


```python
# Let's create a DataFrame with missing values and duplicates
dirty_data = {
    'A': [1.0, 2.0, np.nan, 2.0, 1.0],
    'B': [np.nan, 5.0, 6.0, 5.0, np.nan],
    'C': ['cat', 'dog', 'cat', 'dog', 'cat']
}
df_dirty = pd.DataFrame(dirty_data)
print("Original Dirty DataFrame:")
display(df_dirty)

# Check for Null values
print("\nIsnull mapping:")
display(df_dirty.isnull())

print("\nNull counts per column:")
print(df_dirty.isnull().sum())

# Drop missing values
print("\nDropna (rows with any null):")
display(df_dirty.dropna())

# Fill missing values
print("\nFillna (fill nulls with 0 or mean):")
display(df_dirty.fillna({'A': df_dirty['A'].mean(), 'B': 0}))

# Replace a specific value
print("\nReplace 'cat' with 'feline':")
display(df_dirty.replace('cat', 'feline'))

# Duplicates handling
print("\nDuplicate row check:")
print(df_dirty.duplicated())

print("\nDataFrame after drop_duplicates:")
display(df_dirty.drop_duplicates())

# Rename columns
print("\nRenaming columns:")
display(df_dirty.rename(columns={'A': 'Alpha', 'B': 'Beta'}))
```

## 5. Sorting and Ordering

### Key Methods covered:
*   `df.sort_values(by, ascending)`: Sorts the DataFrame by values in one or more columns.
*   `df.sort_index(ascending)`: Sorts the DataFrame by index labels.


```python
# Sort by Salary descending
print("Sort by Salary (descending):")
display(df.sort_values(by='Salary', ascending=False))

# Sort by City ascending, then Age descending
print("\nSort by City (asc) then Age (desc):")
display(df.sort_values(by=['City', 'Age'], ascending=[True, False]))

# Sorting by index
print("\nSort by index (descending):")
display(df.sort_index(ascending=False))
```

## 6. Data Manipulation and Transformation

Transforming existing columns and applying custom functions.

### Key Methods/Attributes covered:
*   `df.apply(func, axis)`: Applies a function along an axis (rows or columns) of the DataFrame.
*   `df.applymap(func)` / `df.map(func)`: Applies a function element-wise across the entire DataFrame. (Note: `map` is used for element-wise on DataFrames in modern pandas, replacing `applymap`).
*   `series.map(arg)`: Maps values of a Series using input correspondence (dict, series, or function).
*   `df.astype(dtype)`: Casts a pandas object to a specified dtype.
*   `df.assign(new_col=...)`: Assigns new columns to a DataFrame (returns a new DataFrame).
*   `df.drop(labels, axis)`: Removes columns or rows.
*   `series.value_counts()`: Returns counts of unique values.
*   `series.unique()` / `series.nunique()`: Array of unique values and number of unique values.
*   `series.str`: String accessor containing methods like `contains()`, `lower()`, `upper()`, `split()`.


```python
# Let's use apply to double the salaries
print("Using apply to double Salary:")
display(df['Salary'].apply(lambda x: x * 2))

# Map values in a series (e.g. City to country code)
city_map = {'New York': 'USA', 'Paris': 'FRA', 'London': 'GBR', 'Tokyo': 'JPN'}
print("\nMapping City to Country using series.map:")
display(df['City'].map(city_map))

# Change type using astype
print("\nCasting Age to float:")
display(df['Age'].astype(float))

# Adding a column using assign
print("\nAdding a Tax column:")
df_tax = df.assign(Tax = df['Salary'] * 0.15)
display(df_tax)

# Dropping columns
print("\nDropping City and Age:")
display(df.drop(columns=['City', 'Age']))

# Unique values and value counts
print("\nValue counts for City:")
print(df['City'].value_counts())

print("\nUnique cities:", df['City'].unique())
print("Number of unique cities:", df['City'].nunique())

# Using str accessor
print("\nString manipulation (lowercase Name):")
display(df['Name'].str.lower())

print("\nChecking if Name contains 'ar':")
display(df['Name'].str.contains('ar'))
```

## 7. Grouping and Aggregation

The "split-apply-combine" pattern allows you to segment datasets and analyze statistics across categories.

### Key Methods covered:
*   `df.groupby('col')`: Groups DataFrame by a column.
*   `groupby.agg(func)` or `groupby.aggregate(func)`: Compute aggregate statistics.
*   `groupby.mean()`, `groupby.sum()`, `groupby.count()`, `groupby.size()`: Shorthand aggregations.
*   `groupby.transform(func)`: Applies function to groups, returning a Series matching the original size.
*   `groupby.filter(func)`: Filters rows belonging to groups that satisfy a boolean condition.


```python
# Create sample sales dataset
sales_data = {
    'Department': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Electronics', 'Clothing'],
    'Store': ['Store_A', 'Store_B', 'Store_A', 'Store_B', 'Store_A', 'Store_A'],
    'Sales': [1200, 1500, 800, 600, 1100, 750],
    'Quantity': [12, 15, 8, 6, 11, 7]
}
df_sales = pd.DataFrame(sales_data)
print("Sales DataFrame:")
display(df_sales)

# Basic Groupby mean
print("\nMean Sales by Department:")
display(df_sales.groupby('Department')['Sales'].mean())

# Aggregation with multiple functions
print("\nMultiple Aggregations (sum, mean, max) on Sales and Quantity:")
display(df_sales.groupby('Department').agg({
    'Sales': ['sum', 'mean'],
    'Quantity': 'max'
}))

# Groupby transform (returns same shape as original)
print("\nStandardized Sales within each Department (using transform):")
mean_sales = df_sales.groupby('Department')['Sales'].transform('mean')
std_sales = df_sales.groupby('Department')['Sales'].transform('std')
df_sales['Sales_Z_Score'] = (df_sales['Sales'] - mean_sales) / std_sales
display(df_sales)

# Groupby filter (keep groups whose sum of Sales > 2000)
print("\nFiltering to keep only departments with total Sales > 2000:")
display(df_sales.groupby('Department').filter(lambda x: x['Sales'].sum() > 2000))
```

## 8. Combining and Merging DataFrames

Combining separate datasets using concatenation, SQL-like merges, and joins.

### Key Methods/Functions covered:
*   `pd.concat([df1, df2], axis)`: Appends DataFrames along rows (axis=0) or columns (axis=1).
*   `pd.merge(df1, df2, on, how)`: Database-style merge (Inner, Left, Right, Outer).
*   `df1.join(df2)`: Join on index (shorthand for merge).


```python
# DataFrames for merging
df_emp = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Amy', 'Ben', 'Chris', 'David'],
    'Dept_ID': [1, 2, 1, 3]
})

df_dept = pd.DataFrame({
    'Dept_ID': [1, 2, 4],
    'Dept_Name': ['HR', 'IT', 'Marketing']
})

print("Employee DataFrame:")
display(df_emp)
print("Department DataFrame:")
display(df_dept)

# Merge (Inner Join)
print("\nInner Merge on Dept_ID:")
display(pd.merge(df_emp, df_dept, on='Dept_ID', how='inner'))

# Merge (Left Join)
print("\nLeft Merge on Dept_ID:")
display(pd.merge(df_emp, df_dept, on='Dept_ID', how='left'))

# Concatenation (axis=0: stacked vertically)
print("\nVertical concatenation of Employee DataFrames:")
df_emp_new = pd.DataFrame({'Emp_ID': [105], 'Name': ['Emma'], 'Dept_ID': [2]})
display(pd.concat([df_emp, df_emp_new], axis=0, ignore_index=True))
```

## 10. Descriptive Statistics

Pandas provides a suite of math and statistical methods to summarize and evaluate data.

### Key Methods covered:
*   `df.mean()`, `df.median()`, `df.sum()`, `df.min()`, `df.max()`, `df.std()`, `df.var()`: Basic metrics.
*   `df.corr()` / `df.cov()`: Correlation and covariance matrices between numerical columns.
*   `df.cumsum()` / `df.cumprod()`: Cumulative sum and cumulative product.
*   `df.pct_change()`: Percentage change between successive elements.


```python
# Basic math operations
stats_df = pd.DataFrame({
    'X': [10, 15, 20, 25, 30],
    'Y': [2, 4, 8, 16, 32]
})
print("Stats DataFrame:")
display(stats_df)

print("\nSums:\n", stats_df.sum())
print("\nMeans:\n", stats_df.mean())
print("\nMedians:\n", stats_df.median())

# Correlation Matrix
print("\nCorrelation Matrix:")
display(stats_df.corr())

# Cumulative sum
print("\nCumulative Sum of X:")
display(stats_df['X'].cumsum())

# Pct change
print("\nPercentage change in Y:")
display(stats_df['Y'].pct_change())
```
