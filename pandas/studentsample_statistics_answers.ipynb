# Answer Key: 30 Descriptive Statistics Questions
## Mean, Median, Mode, Range, Variance, Std Dev, Quartiles, IQR & Percentiles on `studentsample.csv`

Companion to: `studentsample_statistics_questions.ipynb`

Run the setup cell below before working through the answers.

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("studentsample.csv")

print("Data loaded:", df.shape)
df.head()
```

## Section A: Mean (Questions 1–4)

### 1. Mean age

**Output:** `19.83` (18 non-null ages; S009 and S015 are missing).

```python
df['age'].mean()
```

### 2. Mean GPA (NumPy)

**Output:** `3.34` (18 non-null GPAs; S007 and S017 are missing).

```python
np.mean(df['gpa'].dropna())
```

### 3. Mean attendance percentage

**Output:** `88.72` (18 non-null values).

```python
df['attendance_pct'].mean()
```

### 4. Mean GPA for scholarship students

**Output:** `3.38` (8 students with scholarship `Yes` and non-null GPA).

```python
df[df['scholarship'] == 'Yes']['gpa'].mean()
```

## Section B: Median (Questions 5–7)

### 5. Median age

**Output:** `20.0`

```python
df['age'].median()
```

### 6. Median GPA (NumPy)

**Output:** `3.35`

```python
np.median(df['gpa'].dropna())
```

### 7. Median attendance for male students

**Output:** `85.0` (9 male students with recorded attendance).

```python
df[df['gender'] == 'M']['attendance_pct'].median()
```

## Section C: Mode (Questions 8–10)

### 8. Mode of gender

**Output:** Both `F` and `M` appear 10 times each (bimodal).

```python
df['gender'].mode()
```

### 9. Mode of attendance_pct

**Output:** `90.0` (appears twice; all other values appear once).

```python
df['attendance_pct'].mode()
```

### 10. Mode of scholarship and count

**Output:** Mode is `Yes` with **9** students (excluding 3 missing values).

```python
scholarship_mode = df['scholarship'].mode()[0]
count = (df['scholarship'] == scholarship_mode).sum()
print(f"Mode: {scholarship_mode}")
print(f"Count: {count}")
```

## Section D: Range (Questions 11–13)

### 11. Range of age

**Output:** `4.0` (max 22 − min 18).

```python
df['age'].max() - df['age'].min()
```

### 12. Range of GPA

**Output:** `1.2` (max 3.9 − min 2.7).

```python
df['gpa'].max() - df['gpa'].min()
```

### 13. Range of attendance for female students

**Output:** `11.0` (max 97 − min 86 among 9 female students).

```python
att_f = df[df['gender'] == 'F']['attendance_pct'].dropna()
att_f.max() - att_f.min()
```

## Section E: Variance (Questions 14–16)

### 14. Sample variance of GPA

**Output:** `0.131` (sample variance, ddof=1).

```python
df['gpa'].var()
```

### 15. Sample variance of age (NumPy)

**Output:** `2.029`

```python
ages = df['age'].dropna().values
np.var(ages, ddof=1)
```

### 16. Variance of attendance for 2021 enrollees

**Output:** `34.25` (6 students enrolled in 2021).

```python
df[df['enrollment_year'] == 2021]['attendance_pct'].var()
```

## Section F: Standard Deviation (Questions 17–19)

### 17. Sample std dev of GPA

**Output:** `0.362`

```python
df['gpa'].std()
```

### 18. Sample std dev of attendance (NumPy)

**Output:** `5.475`

```python
att = df['attendance_pct'].dropna().values
np.std(att, ddof=1)
```

### 19. Higher GPA std dev: scholarship Yes or No?

**Output:** Scholarship `Yes` has higher std dev (`0.381` vs `0.298`).

```python
std_yes = df[df['scholarship'] == 'Yes']['gpa'].std()
std_no = df[df['scholarship'] == 'No']['gpa'].std()
print(f"Scholarship Yes: {std_yes:.4f}")
print(f"Scholarship No:  {std_no:.4f}")
print(f"Higher std dev: {'Yes' if std_yes > std_no else 'No'}")
```

## Section G: Quartiles (Questions 20–23)

### 20. First quartile (Q1) of GPA

**Output:** `3.1`

```python
df['gpa'].quantile(0.25)
```

### 21. Third quartile (Q3) of attendance_pct

**Output:** `92.75`

```python
df['attendance_pct'].quantile(0.75)
```

### 22. Second quartile (Q2) of age equals median

**Output:** Q2 = `20.0`, median = `20.0` → they match.

```python
q2 = df['age'].quantile(0.50)
median_age = df['age'].median()
print(f"Q2: {q2}")
print(f"Median: {median_age}")
print(f"Equal: {q2 == median_age}")
```

### 23. All quartiles of GPA

**Output:** Q1=`3.10`, Q2=`3.35`, Q3=`3.60`

```python
df['gpa'].quantile([0.25, 0.50, 0.75])
```

## Section H: Interquartile Range — IQR (Questions 24–26)

### 24. IQR of age

**Output:** `2.0` (Q3 21 − Q1 19).

```python
df['age'].quantile(0.75) - df['age'].quantile(0.25)
```

### 25. IQR of GPA

**Output:** `0.5` (Q3 3.6 − Q1 3.1).

```python
df['gpa'].quantile(0.75) - df['gpa'].quantile(0.25)
```

### 26. IQR of attendance for female students

**Output:** `5.0` (Q3 95 − Q1 90).

```python
att_f = df[df['gender'] == 'F']['attendance_pct'].dropna()
att_f.quantile(0.75) - att_f.quantile(0.25)
```

## Section I: Percentiles (Questions 27–30)

### 27. 90th percentile of GPA

**Output:** `3.8`

```python
df['gpa'].quantile(0.90)
```

### 28. 10th and 75th percentiles of age

**Output:** 10th=`18.0`, 75th=`21.0`

```python
ages = df['age'].dropna().values
p10 = np.percentile(ages, 10)
p75 = np.percentile(ages, 75)
print(f"10th percentile: {p10}")
print(f"75th percentile: {p75}")
```

### 29. 25th percentile of GPA (no scholarship)

**Output:** `3.25` (7 students with scholarship `No`).

```python
df[df['scholarship'] == 'No']['gpa'].quantile(0.25)
```

### 30. Five-number summary of attendance_pct

**Output:** min=`78`, Q1=`85.25`, median=`89.5`, Q3=`92.75`, max=`97`

```python
df['attendance_pct'].describe()
```
