### 60. Stored procedure — CountLowGpaStudents (MySQL)

**Answer:** Returns **3** (S005 gpa=2.8, S011 gpa=2.9, S015 gpa=2.7).

**SQL (MySQL / PostgreSQL):**

```sql
DELIMITER //

CREATE PROCEDURE CountLowGpaStudents(IN p_gpa DECIMAL(3,2))
BEGIN
    SELECT COUNT(*) AS low_gpa_count
    FROM students
    WHERE gpa IS NOT NULL
      AND gpa < p_gpa;
END //

DELIMITER ;

-- Call:
CALL CountLowGpaStudents(3.0);
```

### Bonus: PostgreSQL function (Question 59)

**SQL (MySQL / PostgreSQL):**

```sql
CREATE OR REPLACE FUNCTION get_students_by_major(p_major TEXT)
RETURNS TABLE (
    student_id TEXT,
    name TEXT,
    age DOUBLE PRECISION,
    gender TEXT,
    major TEXT,
    gpa DOUBLE PRECISION,
    attendance_pct DOUBLE PRECISION,
    city TEXT,
    enrollment_year DOUBLE PRECISION,
    scholarship TEXT
) AS $$
BEGIN
    RETURN QUERY
    SELECT s.student_id, s.name, s.age, s.gender, s.major,
           s.gpa, s.attendance_pct, s.city, s.enrollment_year, s.scholarship
    FROM students s
    WHERE s.major = p_major;
END;
$$ LANGUAGE plpgsql;

-- Call:
SELECT * FROM get_students_by_major('Computer Science');
```
